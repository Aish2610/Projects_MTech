```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
dataset = pd.read_csv(r'F:\winequality-red.csv')
```

```python
dataset.shape
```

```python
dataset.plot(x='total sulfur dioxide', y='free sulfur dioxide', style='o')  
plt.title('total sulfur dioxide vs free sulfur dioxide')  
plt.xlabel('total sulfur dioxide')  
plt.ylabel('free sulfur dioxide')  
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.tight_layout()
seabornInstance.distplot(dataset['total sulfur dioxide'])
```

```python
X = dataset['total sulfur dioxide'].values.reshape(-1,1)
y = dataset['free sulfur dioxide'].values.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train) #training the algorithm
```

```python
#To retrieve the intercept:
print(regressor.intercept_)

#For retrieving the slope:
print(regressor.coef_)
```

```python
y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df
```

```python
df1 = df.head(25)
df1.plot(kind='bar',figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='pink')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='blue')
plt.show()
```

```python
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.show()
```
